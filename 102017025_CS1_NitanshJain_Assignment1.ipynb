```python
import numpy as np
```

```python
array1 = np.array([1,2,3,4,5,6])
rev_array1 = np.flip(array1)
print(rev_array1)
```

```python
array1 = np.array([[1,2,3],[2,4,5],[1,2,3]])
flatten_1_array = np.ndarray.flatten(array1)
print(flatten_1_array)
flatten_2_array = np.recarray.flatten(array1)
print(flatten_2_array)
flatten_3_array = np.ravel(array1)
print(flatten_3_array)
```

```python
arr1 = np.array([[1,2],[3,4]])
arr2 = np.array([[1,2],[3,4]])
val = np.array_equal(arr1, arr2)
print(val)
```

```python
x = np.array([1,2,3,4,5,1,2,1,1,1])
val = dict()
for i in x:
  if i not in val:
    val[i]=1
  else:
    val[i]+=1
maxm = max(val, key=val.get)
for i in range(len(x)):
  if x[i]==maxm:
    print(i)
print("Most frequent value = ", maxm)
```

```python
y = np.array([1,1,1,2,3,4,2,4,3,3])
val = dict()
for i in y:
  if i not in val:
    val[i]=1
  else:
    val[i]+=1
maxm = max(val, key=val.get)
for i in range(len(y)):
  if y[i]==maxm:
    print(i)
print("Most frequent value = ", maxm)
```

```python
 gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
 print("Sum of all elements = ",gfg.sum())
 print("Sum of elements row-wise = ", gfg.sum(axis=1))
 print("Sum of elements column-wise = ", gfg.sum(axis=0))
```

```python
n_array = np.array([[55,25,15],[30,44,2],[11,45,77]])
print("Sum of all diagonal elements of the matrix = ", n_array.trace())
print("Eigen values and Eigen Vectors of the matrix = ", np.linalg.eigh(n_array))
print("Inverse of given matrix = ", np.linalg.inv(n_array))
print("Determinant of given matrix", np.linalg.det(n_array))
```

```python
p1 = np.matrix(np.array([[1,2],[2,3]]))
q1 = np.matrix(np.array([[4,5],[6,7]]))
prod_p1_q1 = np.dot(p1,q1)
cov_mat_p1_q1 = np.cov(prod_p1_q1)

p2 = np.matrix(np.array([[1,2],[2,3],[4,5]]))
q2 = np.matrix(np.array([[4,5,1],[6,7,2]]))
prod_p2_q2 = np.dot(p2,q2)
cov_mat_p2_q2 = np.cov(prod_p2_q2)


print("prod_p1_q1 = \n", prod_p1_q1)
print("prod_p2_q2 = \n", prod_p2_q2)
print("cov_mat_p1_q1 = \n", cov_mat_p1_q1)
print("cov_mat_p2_q2 = \n", cov_mat_p2_q2)
```

```python
x = np.matrix([[2,3,4],[3,2,9]])
y = np.matrix([[1,5,0],[5,10,3]])
inner_prod = np.inner(x,y)
outer_prod = np.outer(x,y)
cartesian_prod = np.cross(x,y)

print(inner_prod)
print(outer_prod)
print(cartesian_prod)
```

```python
array = np.array([[1,-2,3],[-4,5,-6]])
abs_arr = np.absolute(array)
flat_arr = np.ravel(array)

print(np.percentile(flat_arr, 25))
print(np.percentile(flat_arr, 50))
print(np.percentile(flat_arr, 75))

print(np.percentile(array, 25, axis=0))
print(np.percentile(array, 50, axis=0))
print(np.percentile(array, 75, axis=0))

print(np.percentile(array, 25, axis=1))
print(np.percentile(array, 50, axis=1))
print(np.percentile(array, 75, axis=1))

print(np.mean(flat_arr))
print(np.median(flat_arr))
print(np.std(flat_arr))

print(np.mean(array, axis=0))
print(np.median(array, axis=0))
print(np.std(array, axis=0))

print(np.mean(array, axis=1))
print(np.median(array, axis=1))
print(np.std(array, axis=1))
```

```python
from PIL import Image
from numpy import asarray

def img_to_array(path, pix_type):
  
  # load the image
  image = Image.open(path).convert(pix_type)

  # convert image to numpy array
  data = asarray(image)
  print(type(data))

  # summarize shape
  print(data.shape)

  return data
```

```python
path = '/content/drive/MyDrive/Colab Notebooks/Machine Learning/images.jpeg'
rgb = 'RGB'
gray_scale = 'L'

data_rgb = img_to_array(path, rgb)
data_grayscale = img_to_array(path, gray_scale)
```

```python
from tabulate import tabulate

datafile_rgb = open('/content/drive/MyDrive/Colab Notebooks/Machine Learning/datafile_rgb.txt', 'a')
datafile_rgb.write(tabulate(data_rgb))
datafile_rgb.close()

datafile_grayscale = open('/content/drive/MyDrive/Colab Notebooks/Machine Learning/datafile_grayscale.txt', 'a')
datafile_grayscale.write(tabulate(data_grayscale))
datafile_grayscale.close()
```
